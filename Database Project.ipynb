```python
import csv

def add():
    no = int(input("Enter the number of students: "))
    
    for i in range(no):
        Firstname = input("Firstname: ").lower()
        Lastname = input("Lastname: ").lower()
        Age = int(input("Age:"))
        CGPA = float(input("CGPA: ")) 
        MAT_NO = input("Matric Number: ").lower()
        clearance = input("Clearance? (Enter 'true' or 'false')").lower() == "true"

        with open("d.csv", "a", newline="") as l:
            writer = csv.DictWriter(l, fieldnames=["firstname", "lastname", "mat_no", "age", "cgpa", "clearance"])
            writer.writerow({
                "firstname": Firstname, 
                "lastname": Lastname, 
                "mat_no": MAT_NO, 
                "age": Age,
                "cgpa": CGPA, 
                "clearance": clearance
            })

def edit():
    students = []
    mat = input("Enter matric number: ").lower()
    key = input("Enter detail to change (firstname, lastname, age, cgpa, clearance): ").lower()

    if key == "age":
        value = int(input("Enter new value: "))
    elif key == "cgpa":
        value = float(input("Enter new value: "))
    else:
        value = input("Enter new value: ")

    with open("d.csv", newline="") as file:
        reader = csv.DictReader(file)
        for row in reader:
            if row["mat_no"] == mat:
                row[key] = value  # Update the row with new value
            students.append(row)  # Keep the updated and original rows
    
    # Overwrite the CSV file with the updated student data
    with open("d.csv", "w", newline="") as file:
        writer = csv.DictWriter(file, fieldnames=["firstname", "lastname", "mat_no", "age", "cgpa", "clearance"])
        writer.writeheader()
        writer.writerows(students)

def display():
    mat = input("Enter matric number: ")
    with open("d.csv", newline="") as file:
        reader = csv.DictReader(file)
        for row in reader:
            if row["mat_no"] == mat:
                print(row)
                break
        else:
            print("Student not found!")

def delete():
    mat = input("Enter matric number of student to delete: ")
    students = []

    with open("d.csv", newline="") as file:
        reader = csv.DictReader(file)
        for row in reader:
            if row["mat_no"] != mat:
                students.append(row)  # Keep students except the one to be deleted

    # Overwrite the CSV with remaining students
    with open("d.csv", "w", newline="") as file:
        writer = csv.DictWriter(file, fieldnames=["firstname", "lastname", "mat_no", "age", "cgpa", "clearance"])
        writer.writeheader()
        writer.writerows(students)
    print(f"Student with Matric Number {mat} has been deleted.")

# Main menu
operation = input("Enter\n1. Add Student\n2. Remove student\n3. Get Information\n4. Edit information\n")

if operation == "1":
    add()
elif operation == "2":
    delete()
elif operation == "3":
    display()
elif operation == "4":
    edit()
```
