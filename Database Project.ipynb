```python
##ADD STUDENT

import csv

def add():
    
    no = int(input("Enter the number of students: "))
    
    for i in range(no):

        Firstname = input("Firstname: ")

        Lastname = input("Lastname: ")

        Age = int(input("Age:"))
        
        CGPA = float(input("CGPA: ")) 

        MAT_NO = input("Matric Number: ")

        clearance = bool(input("Clearance? "))

        with open("d.csv", "a") as l: 
            
            _ = csv.DictWriter(l, fieldnames=["Firstname", "Lastname", "MAT_NO", "Age", "CGPA", "Clearance"])
                        
            _.writerow({"Firstname": Firstname, "Lastname": Lastname, "MAT_NO": MAT_NO, "Age":Age,
                        "CGPA":CGPA, "Clearance":clearance})
            

class Student:
    def __init__(self,firstname, lastname, matric_no,age,cgpa, clearance):
        self.lastname = lastname
        self.firstname = firstname
        self.matric_no  = matric_no
        self.clearance = clearance
        self.age = age
        self.cgpa = cgpa

    def  get_name(self):
        return self.firstname+" "+self.lastname

    def get_age(self):
        return self.age

    def get_clearance(self):
        return self.clearance

    def get_matric_no(self):
        return self.matric_no
    
    def get_age(self):
        return self.age
    
    def change(self, key, value):
        if key == "cgpa":
            self.cgpa = value
        if key == "firstname":
            self.firstname = value
        if key == "lastname":
            self.lastname = value
        if key == "age":
            self.age = value
        if key == "clearance":
            self.clearance = value

def edit():
    with open("d.csv") as file:
        mat = input("Enter matric number: ")
        key = input("Enter detail to change: ")
        key = key.lower()
        if key.lower() == "age":
            value = int(input("Enter new value: "))
        elif key.lower() == "cgpa":
            value = float(input("Enter new value"))
        else:
            value = input("Enter new value: ")
        import csv
        x = csv.DictReader(file)
        for row in x: 
            if row["MAT_NO"] == mat:
                student = Student(row["Firstname"], row["Lastname"], row["MAT_NO"], row["Age"], row["CGPA"], row["Clearance"])
                student.change(key, value)
                
    
def display():
    mat = input("Enter matric number: ")
    with open("d.csv") as file:
        import csv
        x = csv.DictReader(file)
        for row in x:
            if (mat == row["MAT_NO"]):
                print(row.values())
                
            
operation = input("Enter\n1. Add Student\n2. Remove student\n3. Get Information\n4. Edit information\n")

if operation == "1":
    add()
    
elif operation == "2":
    delete()

elif operation == "3":
    display()

elif operation == "4":
    edit()
```
